# Using rank we can compare data in a more covenient way, we can also use these ranks for weighted average.

####  Dataframe for oldest team


```python
df_TeamLaunch_StartingYear=df[['Team','TeamLaunch_StartingYear']]  #.sort_values(by='TeamLaunch_StartingYear')
df_TeamLaunch_StartingYear['TeamLaunch_StartingYear_Rank']=df['TeamLaunch_StartingYear'].rank(method='min')
df_TeamLaunch_StartingYear.sort_values(by='TeamLaunch_StartingYear_Rank').head(10)
```

####  Dataframe for Tournament Champion


```python

df_TournamentChampion = df[['Team','TournamentChampion']]
df_TournamentChampion['TournamentChampion_Rank']=df['TournamentChampion'].rank(method='min',ascending=False)
df_TournamentChampion.sort_values(by=['TournamentChampion_Rank']).head(10)

```

####  Dataframe for percentage of matches won and other dataframes


```python
df_win_per = df[['Team','win_per']]    #win_per- as mentioned above is ( won games/played games)*100
df_win_per['win_per_rank']=df['win_per'].rank(method='min',ascending=False)
df_win_per.sort_values(by=['win_per_rank']).head(10)
```

```python
df_Runnerup = df[['Team','Runner-up']]
df_Runnerup['Runnerup_rank']=df['Runner-up'].rank(method='min',ascending=False)
df_Runnerup.sort_values(by=['Runnerup_rank']).head(10)
```

```python

df_Score_ratio = df[['Team','Score_ratio']]
df_Score_ratio['Score_ratio_rank']=df['Score_ratio'].rank(method='min',ascending=False)
df_Score_ratio.sort_values(by=['Score_ratio_rank']).head(10)
```

```python

df_Tournament = df[['Team','Tournament']]
df_Tournament['Tournament_rank']=df['Tournament'].rank(method='min',ascending=False)
df_Tournament.sort_values(by=['Tournament_rank']).head()
```

```python
df_HighestPositionHeld = df[['Team','HighestPositionHeld']]
df_HighestPositionHeld['HighestPositionHeld_rank']=df['HighestPositionHeld'].rank(method='min')
df_HighestPositionHeld.sort_values(by=['HighestPositionHeld_rank']).head()
```

```python

df['BasketGiven_ratio_rank']=(df['BasketGiven']/df['PlayedGames']).rank(method='min')
df['BasketScored_ratio_rank']=(df['BasketScored']/df['PlayedGames']).rank(method='min',ascending=False)
df['LostGames_ratio_rank']=(df['LostGames']/df['PlayedGames']).rank(method='min')
df['DrawnGames_ratio_rank']=(df['DrawnGames']/df['PlayedGames']).rank(method='min',ascending=False)
```

# For selection of best team we will take the weighted average of all the features we have.
Weighted average is taken as listed below-


Tournament5%	DrawnGamesratio5%	LostGamesratio5%	BasketScoredRatio 5% 	BasketGivenratio 5%	TournamentChampion15% Runner-up10%	HighestPositionHeld 15%	 Score_ratio15%	win_percentage 15%	TeamLaunch_StartingYear 0%

        
           
           ## Formula for weighted average-
             
             
             WeightedAverage=Sum of Weighted Terms/Total Number ofTerms

As some of the older teams are already in contract with competitors, we are not giving weightage to the team launch year we are analysing the other factors.

Weightage can vary according to the preference and opinions.

```python
#Tournament5%	DrawnGamesratio5%	LostGamesratio5%	BasketScored2.5%	BasketGiven2.5%	TournamentChampion15%	Runner-up10%	Tournament5%	DrawnGamesratio5%	LostGamesratio5%	BasketScored2.5%	BasketGiven2.5%	TournamentChampion15%	Runner-up10% HighestPositionHeld15%	Score_ratio20%	win_ratio15%	TeamLaunch_StartingYear 0%
df['Weighted_av']=5*df_Tournament['Tournament_rank']+5*df['BasketGiven_ratio_rank']+5*df['BasketScored_ratio_rank']+5*df['LostGames_ratio_rank']+5*df['DrawnGames_ratio_rank']+15*df_HighestPositionHeld['HighestPositionHeld_rank']+10*df_Runnerup['Runnerup_rank']+0*df_TeamLaunch_StartingYear['TeamLaunch_StartingYear_Rank']+15*df_TournamentChampion['TournamentChampion_Rank']+20*df_Score_ratio['Score_ratio_rank']+15*df_win_per['win_per_rank']
df['Weighted_av']=df['Weighted_av']/5+5+5+5+5+15+10+0+15+20+15
```

```python
df['Weighted_av_rank']=df['Weighted_av'].rank(method='min')
```

```python
df_BestTeam = df[['Team','Weighted_av_rank']]
```
